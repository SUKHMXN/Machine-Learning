```python
df=pd.read_csv('data_stress.csv')
df.head(3)
```

```python
df.drop(['blood oxygen ','limb movement','eye movement'], axis=1,inplace=True)
```

```python
df.head(2)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df.hist(bins=50,figsize=(20,15))
```

```python
sns.boxplot(df)
```

```python
for i in df.columns:
    fig, axes = plt.subplots(1,1)
    sns.distplot(df[i], kde = True, color ='red', bins = 30)
```

```python
df.info()
```

```python
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan, strategy='mean')
imp.fit_transform([df['body temperature']]).shape
mean_val=df['body temperature'].mean()
mean_sleep=df['hours of sleep'].mean()
mean_heartrate=df['heart rate '].mean()
df1=df['body temperature']
df2=df['hours of sleep']
df3=df['heart rate ']
df['body temperature'].fillna(mean_val,inplace=True)
df['hours of sleep'].fillna(mean_sleep,inplace=True)
df['heart rate '].fillna(mean_heartrate,inplace=True)
# None in pd.unique(df['body temperature'])
```

```python
sns.distplot(df['heart rate '], kde = True, color ='red', bins = 30,hist=False)
sns.distplot(df3, kde = True, color ='blue', bins = 30,hist=False)
```

```python
sns.distplot(df['body temperature'], kde = True, color ='red', bins = 30,hist=False)
sns.distplot(df1, kde = True, color ='blue', bins = 30,hist=False)
```

```python
sns.distplot(df['hours of sleep'], kde = True, color ='red', bins = 30)
sns.distplot(df2, kde = True, color ='blue', bins = 30,hist=False)
```

```python
from sklearn.neighbors import KNeighborsClassifier
pipe = Pipeline([('scaler', StandardScaler()),('KNN',KNeighborsClassifier(n_neighbors=4))])
```

```python
X=df.drop('Stress Levels',axis=1)
y=df['Stress Levels']
df.info()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,random_state=0)
pipe.fit(X_train, y_train)
pipe.named_steps
```

```python
y_pred=pipe.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(pipe,X_train,y_train,cv=5,scoring='accuracy').mean()
```
